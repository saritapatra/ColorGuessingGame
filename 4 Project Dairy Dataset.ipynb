```python
import pandas as pd
```

```python
df = pd.read_csv("dairy_dataset.csv")
df
```

```python
df = df.drop(["Farm Size","Product ID","Sales Channel","Reorder Quantity (liters/kg)"], axis = 1)
df
```

##1. does number of cows in a farm has any dependency on area aquired. 

```python
farm = df.groupby("Location").agg({"Total Land Area (acres)":"sum","Number of Cows":"sum" })
farm
```
